# Abstract

The information we have here is part of a study focused on a liver condition called non-alcoholic fatty liver disease (NAFLD). This happens when the liver stores too much fat, similar to what can occur in people who are overweight or obese. NAFLD isn't just a harmless issue; it can lead to more severe liver problems, such as non-alcoholic steatohepatitis (NASH), which can cause serious liver damage.

Now, researchers are using this dataset like a roadmap to understand NAFLD better. They're interested in figuring out how different factors, like a person's age, weight, and whether they're male or female, might influence how NAFLD develops over time. By collecting and analyzing this information, scientists hope to uncover patterns or trends that could provide valuable insights into the progression of NAFLD. Ultimately, this knowledge could lead to more effective treatments and interventions for people living with this condition.

## Understanding Attributes

1) ID: This is a unique number assigned to each person participating in the study. It helps researchers keep track of individuals and their data throughout the study period.
2) Age: Age refers to how many years old each person was when they started participating in the study. Age can be an important factor in understanding how health conditions develop and progress over time.
3) Gender: Gender indicates whether a person is male or female. This demographic information is often considered in medical research because certain health conditions may affect men and women differently.
4) Weight: Weight is a measure of how much mass an individual carries, typically measured in kilograms. In this study, weight data is collected to understand the relationship between body weight and the development or progression of non-alcoholic fatty liver disease (NAFLD).
5) Height: Height refers to how tall a person is, measured in centimeters. Height, when combined with weight, is used to calculate the Body Mass Index (BMI), which helps assess whether a person's weight is appropriate for their height.
6) BMI (Body Mass Index): BMI is a calculated value derived from a person's weight and height. It provides an indication of whether a person's weight is within a healthy range for their height. Higher BMI values typically correlate with increased risk for health issues like NAFLD.
7) Case ID: Each participant in the study is matched with a specific case of non-alcoholic fatty liver disease (NAFLD). This linkage allows researchers to analyze how individual characteristics relate to the presence and progression of NAFLD.
8) Futime: This attribute represents the time elapsed from the beginning of the study until a specific event occurs, which could be the participant's death or their last follow-up in the study. 
9) Status: Status indicates whether a person was alive or deceased by the end of the study period. This information helps researchers understand the impact of NAFLD and related factors on mortality rates within the study population.


Columns	Description

## Data Preparation

```python
# fills NA/NaN values
df['weight'].fillna(df['weight'].median(), inplace=True)
```

```python
# fills NA/NaN values
df['height'].fillna(df['height'].median(), inplace=True)
```

```python
# fills NA/NaN values
df['bmi'].fillna(df['bmi'].median(), inplace=True)
```

```python
# drops columns from the data frame
df.drop('Unnamed: 0', axis=1, inplace=True)
df.tail()
```

```python
# drops columns from the data frame
df.drop('case.id', axis=1, inplace=True)
df.tail()
```

```python
# drops columns from the data frame
df.drop('id', axis=1, inplace=True)
df.tail()
```

```python
# import StandardScaler
from sklearn.preprocessing import StandardScaler
scal=StandardScaler()
feat=['age', 'male','weight', 'height', 'bmi','futime']
# fits to data and transforms it
df[feat] = scal.fit_transform(df[feat])
# shows the first five column of the data frame
df.head()
```

**Creating Features and Target variable**

```python
# drop target variable status from the feature set
X=df.drop("status",axis=1).values
Y=df.status.values
```

**Splitting the data into train and test sets**

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split

# Define X and y using entire dataset


# Define the desired ratio for the minority class after oversampling
desired_ratio = 0.30

# Define the ratio for undersampling the majority class
undersampling_ratio = 1  # For example, if you want the majority class to be half the size of the minority class

# Define the pipeline
pipeline = Pipeline([
    ('oversample', SMOTE(sampling_strategy=desired_ratio, random_state=42)),
    ('undersample', RandomUnderSampler(sampling_strategy=undersampling_ratio, random_state=42))
    
])

# Apply the pipeline to resample the entire dataset
X_resampled, y_resampled = pipeline.fit_resample(X, Y)

# Split the resampled data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)


```

**Create a function for evaluating metrics**

```python
from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score,roc_auc_score,confusion_matrix

def evaluation(Y_test,Y_pred):
    # accuracy classification score
    acc=accuracy_score(Y_test,Y_pred)
    # computes recall
    rcl=recall_score(Y_test,Y_pred)
    # computes f1 score
    f1=f1_score(Y_test,Y_pred)
    
    metric_dict={'accuracy': round(acc,3),
                 'recall': round(rcl,3),
                 'F1 score': round(f1,3), 
              }
    
    return print(metric_dict)
```

## Building a Model

**Fitting and Comparing different Models**

```python
np.random.seed(42)
from sklearn.neighbors import KNeighborsClassifier
Knn_clf=  KNeighborsClassifier()
Knn_clf.fit(X_train,Y_train)
Knn_Y_pred=Knn_clf.predict(X_test)
Knn_score=Knn_clf.score(X_test,Y_test)
#print(Knn_score)
evaluation(Y_test,Knn_Y_pred)
```

```python
np.random.seed(42)
from sklearn.linear_model import LogisticRegression
LR_clf=LogisticRegression()
LR_clf.fit(X_train,Y_train)
LR_Y_pred=LR_clf.predict(X_test)
LR_score=LR_clf.score(X_test,Y_test)
#print(LR_score)
evaluation(Y_test,LR_Y_pred)
```

```python
np.random.seed(42)
from sklearn.ensemble import RandomForestClassifier
RF_clf=RandomForestClassifier(n_estimators=450)
RF_clf.fit(X_train,Y_train)
RF_score=RF_clf.score(X_test,Y_test)
RF_Y_pred=RF_clf.predict(X_test)
#print(RF_score)
evaluation(Y_test,RF_Y_pred)
```

```python
np.random.seed(42)
from sklearn.svm import SVC
SVC_clf=SVC()
SVC_clf.fit(X_train,Y_train)
SVC_score=SVC_clf.score(X_test,Y_test)
SVC_Y_pred=SVC_clf.predict(X_test)
#print(SVC_score)
evaluation(Y_test,SVC_Y_pred)
```

```python
%%writefile app.py
import streamlit as st
import base64
import sklearn
import pandas as pd
import numpy as np
import random
from joblib import load
from sklearn.preprocessing import StandardScaler


svcmodel = load('SVC_model.joblib')
xgbmodel=load('XGB_model.joblib')
scal = load('scal_model.joblib')
models=[svcmodel,xgbmodel]
model=random.choice(models)
st.set_page_config(
    page_title="Non-alcohol fatty liver disease prediction",
    page_icon="",
    layout="centered",
    initial_sidebar_state="expanded",
    menu_items={
        'About': "This is an app to predict your risk of getting non-alcohol fatty liver disease!"
    })

def preprocess(user_input):  
    # Pre-processing user input   
    user_input = pd.DataFrame([user_input],columns=['age','male','weight','height','bmi','futime'])
    user_input=scal.fit_transform(user_input)
    
    return user_input

# front end elements of the web page 
html_temp = """ 
    <div> 
    <h2 style ="color:black;text-align:center;">Non-alcohol fatty liver disease prediction</h2> 
    </div> 
    """
      
st.markdown(html_temp, unsafe_allow_html = True) 

age=st.selectbox ("Age",range(1,100,1))
male = st.radio("Select sex: ", ('male', 'female'))
weight=st.selectbox('Weight in kg',range(1,300,1))
height=st.selectbox('Height in cm',range(1,200,1))
bmi=st.selectbox('BMI',range(1,50,1))
futime=st.selectbox ("Time to death or last follow up",range(1,10000,1))
if male=="male":
    male=1 
else: 
    male=0 
pred=model.predict(preprocess([age,male,weight,height,bmi,futime]))
print(preprocess([age,male,weight,height,bmi,futime]))

if st.button("Predict"):    
    if pred[0] == 1:
        st.error('Warning! You have high risk of dying from fatty liver!')
    else:
        st.success('You have lower risk of dying from fatty liver!')
        
st.sidebar.subheader("About")

st.sidebar.info("This web application helps you to find out whether you are at a risk of dying from fatty liver.")
st.sidebar.info("Enter the required fields and click on the 'Predict' button to check whether you have a healthy liver")

st.info("Caution: This is just a prediction and not doctoral advice.") 
```
